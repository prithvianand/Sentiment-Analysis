```python
# this will check for the datatype of the review data of every row.
# if any row has different data type it will convert it to string
for row_value  in range(0,len(feedbacks)-1):
    if type(feedbacks.iloc[row_value]['Response']) != str:
        feedbacks.iloc[row_value]['Response'] = str(feedbacks.iloc[row_value]['Response'])
```

```python
from textblob import TextBlob
```

```python
list_senti = []
list_pol = []
list_sub = []

for feedback in feedbacks['Response']:
    blob = TextBlob(feedback)
    senti = blob.sentiment
    pol = blob.sentiment.polarity
    sub = blob.sentiment.subjectivity
    print(senti)
    list_senti.append(senti)
    list_pol.append(pol)
    list_sub.append(sub)
```
